```python
from langchain_core.tools import Tool
from pydantic import BaseModel, Field

class TemperatureInput(BaseModel):
    fahrenheit: float = Field(..., description="Temperature in Fahrenheit")

def fahrenheit_to_celsius(fahrenheit: float) -> str:
    celsius = (fahrenheit - 32) * 5 / 9
    return f"{celsius:.2f}°C"

def temperature_conversion_func(input_str: str) -> str:
    try:
        fahrenheit_value = float(input_str)
        return fahrenheit_to_celsius(fahrenheit_value)
    except ValueError:
        return "Invalid input. Please provide a numeric value for Fahrenheit temperature."

temperature_conversion_tool = Tool(
    name="TemperatureConversion",
    func=temperature_conversion_func,
    description="Converts temperature from Fahrenheit to Celsius."
)
```

```python
# Create the tool
tools = [temperature_conversion_tool,
    Tool(
        name="web search",
        func=search.run,
        description="Useful when you want to fetch information from the web for questions where you dont have the answer"
    )
]
```

```python
# Create the ReAct agent
agent = create_react_agent(llm=llm, tools=tools, prompt=prompt)

# Wrap the agent in an executor
agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
while True:
  template_str = "What is the temperature of {city}. Provide the temperature in celcius."
  user_input = input("Enter a city name to know its temperature: ")
  if user_input.lower() == "exit":
    break
  template_str = template_str.format(city=user_input)
  response = agent_executor.invoke({"input":template_str})
  print(response["output"])
```
